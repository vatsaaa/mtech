````python
# @title Import required libraries
import pandas as pd
import re
import torch
import spacy
from transformers import AutoModelForTokenClassification, AutoTokenizer, pipeline
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report
from tqdm import tqdm
````

````python
# @title Load dataset
def load_dataset(file_path):
    try:
        df = pd.read_csv(file_path)
        return df
    except FileNotFoundError:
        print(f"Error: File '{file_path}' not found.")
        return None
    except Exception as e:
        print(f"An error occurred while loading the dataset: {e}")
        return None
````

````python
# @title Preprocess dataset: Tokenization, cleaning, and formatting for model training
def preprocess_data(df):
    df.dropna(subset=['Filled Template', 'Tokens'], inplace=True)
    df['Tokens'] = df['Tokens'].apply(lambda x: eval(x))
    return df

# @title Preprocess dataset
def preprocess_data(df):
    if df is None:
        return None

    df.dropna(subset=['Filled Template', 'Tokens'], inplace=True)
    df['Tokens'] = df['Tokens'].apply(lambda x: eval(x))
    return df
````

````python
# @title Load pre-trained NER model
def load_ner_model():
    model_name = "dbmdz/bert-large-cased-finetuned-conll03-english"
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = AutoModelForTokenClassification.from_pretrained(model_name)
    ner_pipeline = pipeline("ner", model=model, tokenizer=tokenizer)
    return ner_pipeline
````

# Embedding Techniques Used
BERT embeddings are loaded and used in load_ner_model() function, when using BertForTokenClassification and the BERT tokenizer

## WordPiece Tokenization (Subword Embeddings)
- The BERT tokenizer (BertTokenizerFast) is used to split words into subwords or wordpieces (e.g., "playing" → ["play", "##ing"]).
- This ensures better handling of rare words and out-of-vocabulary (OOV) tokens.

## Contextualized BERT Embeddings
- We use AutoTokenizer and AutoModelForTokenClassification for model flexibility and future compatibility:
  - **Model Flexibility:** AutoTokenizer and AutoModelForTokenClassification allow switching between different transformer models (BERT, RoBERTa, DeBERTa, etc.) without changing the code.
    
    If later we decide to use a different model like roberta-base, you only need to change model_name.

  - **Future Compatibility:** The Auto* classes automatically select the correct architecture based on model_name, ensuring you don’t manually pick the wrong tokenizer/model.

- BERT (bert-base-uncased) converts tokenized input into dense, contextualized embeddings.
- Unlike static embeddings (e.g., GloVe, Word2Vec), BERT generates different embeddings for the same word based on context.
- Example:
  - "He went to the bank to withdraw money." → "bank" (financial institution)
  - "She sat on the river bank." → "bank" (side of a river)

## Pre-trained BERT Model for Token Classification
- The model uses BertForTokenClassification, which is a fine-tuned BERT model specifically designed for sequence labeling tasks like NER.
- Each token's embedding is passed through a classification layer to predict entity labels (PER, ORG, LOC, etc.).

#  Why did we use BERT?
- Handles polysemy (multiple meanings of words)
- Contextual embeddings improve NER accuracy
- Works well even with small datasets (transfer learning)

````python
# @title Extract named entities
def extract_named_entities(text, ner_pipeline):
    entities = ner_pipeline(text)
    return entities
````

````python
# @title Mask sensitive information
def mask_text(text, entities):
    masked_text = text
    masked_indices = set()

    for entity in entities:
        if entity['entity'].startswith('B-') or entity['entity'].startswith('I-'):
            start_index = entity['start']
            end_index = entity['end']
            for i in range(start_index, end_index):
                masked_indices.add(i)

    masked_chars = []
    for i, char in enumerate(text):
        if i in masked_indices:
            masked_chars.append('x')
        else:
            masked_chars.append(char)

    masked_text = ''.join(masked_chars)

    # Custom masking for phone numbers and emails
    masked_text = re.sub(r'\b\+?\d{10,15}\b', lambda match: 'x' * len(match.group(0)), masked_text)
    masked_text = re.sub(r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,7}\b', lambda match: 'x' * len(match.group(0)), masked_text)

    return masked_text
````

````python
# @title Evaluate model performance
def evaluate_model(test_data, ner_pipeline):
    if test_data is None:
        return

    y_true = []
    y_pred = []

    for text, true_labels in tqdm(zip(test_data['Filled Template'], test_data['Tokens']), total=len(test_data)):
        predicted_entities = extract_named_entities(text, ner_pipeline)
        predicted_labels = ['O'] * len(true_labels)
        text_split = text.split()
        for entity in predicted_entities:
            try:
                word = entity['word']
                if word in text_split:
                    idx = text_split.index(word)
                    predicted_labels[idx] = entity['entity']
            except ValueError:
                continue
        y_true.extend(true_labels)
        y_pred.extend(predicted_labels)

    print(classification_report(y_true, y_pred, zero_division=0))
    cm = confusion_matrix(y_true, y_pred, labels=list(set(y_true)))
    plt.figure(figsize=(12, 8))
    sns.heatmap(cm, annot=True, fmt='d', xticklabels=list(set(y_true)), yticklabels=list(set(y_true)))
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.title("Confusion Matrix")
    plt.show()
````

````python
# @title Main function
def main():
    dataset_path = "Dataset.csv"
    df = load_dataset(dataset_path)
    df = preprocess_data(df)

    if df is not None:
        ner_pipeline = load_ner_model()

        # Show a few examples from the dataset
        print("Printing some examples from the dataset:")
        for i in range(min(5, len(df))):
            original_text = df['Filled Template'].iloc[i]
            entities = extract_named_entities(original_text, ner_pipeline)
            masked_output = mask_text(original_text, entities)
            print(f"\nOriginal: {original_text}")
            print(f"Masked: {masked_output}")

        # User input loop
        while True:
            user_input = input("\nEnter text to mask (or type 'exit' to quit): ")
            if user_input.lower() == 'exit':
                break

            entities = extract_named_entities(user_input, ner_pipeline)
            masked_output = mask_text(user_input, entities)
            print(f"Masked Output: {masked_output}")

        evaluate_model(df, ner_pipeline)
````

````python
main()
````
