# Linear Systems



## Example 1: Solve the following system of equations

$x_1 + 3x_2 = 5$

$x_1 - 2x_2 = -5$

## Solution:

We have two lines here, plot each and see where these intersect to get a solution

$x_2 = \frac{5\ -\ x_1} {3}$


Let us plot for this now, then we will plot for

$x_2 = \frac{5 + x_1} {2}$

```python
import numpy as np
import matplotlib.pyplot as plt

# Plotting line1
x1 = np.linspace(-10, 10, 100)
y1 = (5 - x1)/3

fig, ax = plt.subplots()
ax.plot(x1, y1)
ax.text(1,1.6,'$x_1+3x_2 = 5$')

# Plotting line2 but on the same figure and axes
x2 = np.linspace(-10, 10, 100)
y2 = (5 + x2)/2

# Since it is on the same figure and axes, we do not call subplots() again
ax.plot(x2, y2)
ax.text(-3,0.5,'$x_1-2x_2 = -5$')

# Solve the 2 equations
A = np.array([[1, 3], [1, -2]])
B = np.array([5, -5])

X = np.linalg.solve(A, B)

# Display the solution
ax.plot(X[0], X[1], 'ro')
```

## 3: Consider the system of equations given below. Under what conditions on the coefficients $a, b, c, d, e, f$ does this system have only one solution? No solutions? An infinite number of solutions? (Think about what must be true of the slopes and intercepts of the lines in each case.)

$ax_1+bx_2 = c$

$dx_1+ex_2=f$



Solution:
A system has
- only one solution when $\frac{ùëé} {d}$ is not equal to $\frac{b} {e}$
- no solution when $\frac{a} {d} = \frac{b} {e} \neq \frac{c} {f}$
- an infinite number of solutions when $\frac{a} {d} = \frac{b} {e} = \frac{c} {f}$
